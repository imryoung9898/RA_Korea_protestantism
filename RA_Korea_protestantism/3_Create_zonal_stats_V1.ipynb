### Create zonal statistics
This code creates zonal statistics from raster (e.g. Nightlights) <br>
Last modified : Apr 7 2021 by Imryoung Jeong (neptune0118@gmail.com)


```python
import geopandas as gpd
import rasterio as rio
import numpy as np
import rasterstats
import pandas as pd
import os, re
import re

pd.set_option('display.float_format', '{:.5f}'.format)
```

```python
dir_shp = "/Users/imryoung/Dropbox/EconAI_Maps/Data/cookie_mold/Population_grid/population_grid_combined_2km.shp"
dir_raster = "/Users/imryoung/Dropbox/Nightlight_global/Harmonized_92_18/"
dir_out = "/Users/imryoung/Dropbox/EconAI_Imryoung/Data/Nightlight_2016.csv"

mask = gpd.read_file(dir_shp, encoding = 'euc-kr').set_crs(5179).to_crs(4326)
mask.head(3)
```

```python
raster = rio.open(dir_raster + "Harmonized_DN_NTL_2016_simVIIRS.tif")
raster_array = raster.read(1)
affine = raster.transform

avg = rasterstats.zonal_stats(mask, raster_array, 
                             affine = affine, stats = ['mean'], geojson_out = True)

avg_nightlight = []
i=0

while i < len(avg):
    avg_nightlight.append(avg[i]['properties'])
    i += 1
        
df = pd.DataFrame(avg_nightlight)
df.to_csv(dir_out)
```
